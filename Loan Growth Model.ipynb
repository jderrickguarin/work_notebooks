### Tests for Stationarity

```python
#testing for stationarity
def adf_test(series, name=None, num=None):
    result = adfuller(series)
    if name == None:
        name = str(series.name)
    if num != None:
        name = 'Series %s' % str(num)
    print('\n%s:' %name)
    print('ADF Statistic: %f' % result[0])
    print('p-value: %f' % result[1])
    
for s in series:
    adf_test(s)
```

#### Correcting Heteroskedasticity

```python
# correcting increasing volatility

def annual_vol(series):
    stdevs = series.groupby(series.index.year).std()
    return series.index.map(lambda d: stdevs.loc[d.year])

def variance_correct(series):
    return series/annual_vol(series)

gdp = variance_correct(gdp)
loans = variance_correct(loans)
gdp_growth = variance_correct(gdp_growth)
loans_growth = variance_correct(loans_growth)
ipi_vol = variance_correct(ipi_vol)
ipi_val = variance_correct(ipi_val)
pmi = variance_correct(pmi)
rr_rate = variance_correct(rr_rate)
```

#### Removing Seasonality

```python
# correcting seasonality

def qtr_mean(series):
    means = series.groupby(series.index.month).mean()
    return series.index.map(lambda d: means.loc[d.month])

def seasonality_correct(series):
    return series - qtr_mean(series)

gdp = seasonality_correct(gdp)
loans = seasonality_correct(loans)
gdp_growth = seasonality_correct(gdp_growth)
loans_growth = seasonality_correct(loans_growth)
ipi_vol = seasonality_correct(ipi_vol)
ipi_val = seasonality_correct(ipi_val)
pmi = seasonality_correct(pmi)
rr_rate = seasonality_correct(rr_rate).dropna()
```

### Running Granger Causality on Corrected Data

```python
def granger(s1, s2, lag=4):
    granger_df = pd.concat([s1, s2], axis=1).dropna()
    grangercausalitytests(granger_df, lag)
    
comb = [[loans_growth, gdp_growth],
       [loans_growth, ipi_val],
       [loans_growth, ipi_vol],
       [loans_growth, pmi],
       [loans_growth, rr_rate],
       [loans, gdp_growth],
       [loans, ipi_val],
       [loans, ipi_vol],
       [loans, pmi],
       [loans, rr_rate]]
```

```python
#try on every combination
num = 0
for c in comb:
    num += 1
    print('Granger Causality Results for Relation %s' %str(num))
    print('-------------------------------------------------')
    print('\nBetween %s' %str(c[0].name)+' and '+str(c[1].name))
    print('\n%s as Affected: ' %str(c[0].name))
    granger(c[0], c[1], lag=4)
    print('\n%s as Affected: ' %str(c[1].name))
    granger(c[1], c[0], lag=4)
    print('\n=================================================\n')
```

#### Summary

The results showed that, considering that p=0456, **the GDP growth rate Granger causes UKB loans** at lag 3, but Loans do not Granger cause GDP growth.

Interestingly, **y-o-y loans growth Granger causes the Industrial Production Index (IPI) Volume** with 1 and 2 lags (p=0.0382, p=0.0382). Meaning, there is statistical evidence to say that rate of growth of loans one or two months prior may Granger cause the IPI Volume.

```python
### AUTOMATE MODEL SELECTION
def fit_tsmodel(data, model='VAR', lags=4):
    models = ['VAR','SARIMAX']

    if model in models:
        if model == models[0]:
            model = VAR
        elif model == models[1]:
            model == SARIMAX
    else:
        print('Model not supported')
        return None
    
    fmodel = model(data)          
    res = fmodel.fit(maxlags=lags)
    return res

def create_df(*series):
    df = pd.concat(series, axis=1).dropna()
    n = []
    for i in series:
        n.append(i.name)
    df.columns = n
    return df

##---test---
#df1 = create_df(loans_growth, ipi_vol)
#fit_tsmodel(df1, 'VAR').summary()

    
```

```python
def check_combinations(classifiers, data_slice, transformation, X_train, y_train, X_test, y_test):
    model_names = []
    data_slice = [data_slice] * len(classifiers)
    transformation = [transformation] * len(classifiers)
    train_accuracies = []
    train_positive_recalls = []
    test_accuracies = []
    
    for clf in classifiers:
        clf.fit(X_train, y_train)
        model = str(type(clf)).split(".")[-1][:-2]

        y_pred = clf.predict(X_train)
        report = classification_report(y_train, y_pred, target_names=target_names, output_dict=True)
        train_accuracy = report['accuracy']
        train_positive_recall = report['Positive']['recall']

        y_pred = clf.predict(X_test)
        report = classification_report(y_test, y_pred, target_names=target_names, output_dict=True)
        test_accuracy = report['accuracy']
        
        model_names.append(model)
        train_accuracies.append(train_accuracy)
        train_positive_recalls.append(train_positive_recall)
        test_accuracies.append(test_accuracy)
    
    dict = {'model_names': model_names, 'data_slice': data_slice, 
            'transformation': transformation, 'train_accuracies': train_accuracies, 
            'train_positive_recalls': train_positive_recalls, 'test_accuracies': test_accuracies}  
    
    df = pd.DataFrame(dict)
    return df
```
